```python
#Replace blank cells in the "Flouride" column with "0"
df["Fluoride (mg/L)"]=df["Fluoride (mg/L)"].fillna(0)
df["Fluoride (mg/L)"]
```

```python
#Filter the dataset to only show data for the year 2021
df= df[df["Sample Date"].str.contains("2021")]
df
```

```python
#Export changes to a new CSV file
df.to_csv('./DrinkingWaterQuality2021Cleaned.csv')
```
